```python
#Imports
import pandas as pd
import numpy as np
```

```python
#Read the csv files
rated_df = pd.read_csv("olivia_work.csv")
rated_df
```

```python
#Delete other string columns
del rated_df['Unnamed: 0']
del rated_df['Title']
del rated_df['Country']
del rated_df['Director']
del rated_df['Production']
```

```python
rated_df
```

```python
rated=[]

for i in np.array(rated_df['Rated'].apply(lambda x: x.split(','))):
    for j in i:
        rated.append(j.strip())
```

```python
set(rated)
```

```python
# define columns first
for i in set(rated):
    rated_df[i]= pd.Series()

# populating colmuns 
for x in rated_df['Rated']:
    print(x.split(','))
```

```python
rated_df['R'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(rated_df['Rated']):
    
    for words in x.split(','):
        if words.strip() in set(rated):
            rated_df[words.strip()][i] = 1
```

```python
rated_df.fillna(0, inplace=True)

del rated_df['Rated']
```

```python
rated_df = rated_df.apply(lambda x: pd.to_numeric(x), axis=0)
rated_df = rated_df.astype(int)
rated_df.to_csv('ratings_integers_only.csv')

```
