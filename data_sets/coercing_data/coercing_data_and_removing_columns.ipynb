```python
from sklearn.datasets import make_blobs
# read file
df = pd.read_csv("../combined_encoded_final_updated.csv")
```

```python
df.columns
```

```python
# Dropping columns that confuse regression - unnamed, released, box office, Domestic Box Office, International Box Office
df.drop(["Unnamed: 0"], axis=1, inplace=True)
df.drop(["Released"], axis=1, inplace=True)
df.drop(["Box Office"], axis=1, inplace=True)
df.drop(["Domestic Box Office"], axis=1, inplace=True)
df.drop(["International Box Office"], axis=1, inplace=True)
```

```python
# renaming updated_df as df1
df1 = df.fillna(0)
df1.head()
```

```python
# Coercing data
df1["Worldwide Box Office"] = pd.to_numeric(df1["Worldwide Box Office"], errors='coerce')
df1["IMDB"] = pd.to_numeric(df1["IMDB"], errors='coerce')
df1["Rank"] = pd.to_numeric(df1["Rank"], errors='coerce')
df1["Metacritic"] = pd.to_numeric(df1["Metacritic"], errors='coerce')
df1["Rotten Tomatoes"] = pd.to_numeric(df1["Rotten Tomatoes"], errors='coerce')
df1["Worldwide Box Office"].head()
```
