```python
#Imports
import pandas as pd
import numpy as np
```

```python
#Read the csv files
oscars_df = pd.read_csv("Resources/Oscar_Awards.csv")
boxoffice_df = pd.read_csv("Resources/movie_list_all_backup.csv")
all_movies_df = pd.read_csv("Resources/Movie_DF_2.csv")
```

```python
#Inner merge the two csv files 
movies_boxoffice_df = pd.merge(all_movies_df, boxoffice_df, how='inner')
movies_boxoffice_df.head()
movies_boxoffice_df.to_csv('movies_boxoffice.csv')
```

```python
#Left merge the two ccsv files 
combined_df = pd.merge(movies_boxoffice_df, oscars_df, how='left')
combined_df.dtypes
```

```python
#Replace Na's with 0
combined_df.fillna(0, inplace=True)

#Remove $ and , from columns 
combined_df["Worldwide Box Office"] = combined_df["Worldwide Box Office"].map(lambda x: x.lstrip('$-').rstrip('aAbBcC')).str.replace(r'[^\d.]+', '')
combined_df["Domestic Box Office"] = combined_df["Domestic Box Office"].map(lambda x: x.lstrip('$-').rstrip('aAbBcC')).str.replace(r'[^\d.]+', '')
combined_df["International Box Office"] = combined_df["International Box Office"].map(lambda x: x.lstrip('$-').rstrip('aAbBcC')).str.replace(r'[^\d.]+', '')
combined_df["Rank"] = combined_df["Rank"].str.replace(r'[^\d.]+', '')

#Remove unwanted symbols at the end of values 
combined_df["IMDB"] = combined_df["IMDB"].map(lambda x: str(x)[:-3])
combined_df["Release Year"] = combined_df["Release Year"].map(lambda x: str(x)[:-2])
combined_df["Awards"] = combined_df["Awards"].map(lambda x: str(x)[:-2])
combined_df["Nominations"] = combined_df["Nominations"].map(lambda x: str(x)[:-2])
combined_df["Box Office"] = combined_df["Box Office"].map(lambda x: str(x)[:-2])
combined_df["Metacritic"] = combined_df["Metacritic"].map(lambda x: str(x)[:-4])
combined_df["Runtime"] = combined_df["Runtime"].map(lambda x: str(x)[:-4])
combined_df["Rotten Tomatoes"] = combined_df["Rotten Tomatoes"].map(lambda x: str(x)[-2:])

#Delete unwanted columns 
del combined_df['Year']
del combined_df['Actors']
del combined_df['Writer']
del combined_df['Unnamed: 0']
del combined_df['Type']

combined_df
```

```python
#Converting column values to integers
combined_df["Worldwide Box Office"] = combined_df["Worldwide Box Office"].astype(str).astype(int)
combined_df["Domestic Box Office"] = combined_df["Domestic Box Office"].astype(str).astype(int)
combined_df["International Box Office"] = combined_df["International Box Office"].astype(str).astype(int)
combined_df["Box Office"] = combined_df["Box Office"].astype(int)
combined_df["Released"] = combined_df["Released"].astype(int)
combined_df["Release Year"] = combined_df["Release Year"].astype(int)
combined_df["Awards"] = combined_df["Awards"].astype(int)
combined_df["Nominations"] = combined_df["Nominations"].astype(int)

combined_df["IMDB"] = pd.to_numeric(combined_df["IMDB"])
combined_df["Metacritic"] = pd.to_numeric(combined_df["Metacritic"])
combined_df["Rank"] = pd.to_numeric(combined_df["Rank"])
combined_df["Rotten Tomatoes"] = pd.to_numeric(combined_df["Rotten Tomatoes"])
combined_df["Runtime"]= pd.to_numeric(combined_df["Runtime"])


combined_df["IMDB"] = combined_df["IMDB"].astype(int)
combined_df["Metacritic"].fillna(0, inplace=True)
combined_df["Metacritic"] = combined_df["Metacritic"].astype(int)
combined_df["Rank"] = combined_df["Rank"].astype(int)
combined_df["Rotten Tomatoes"] = combined_df["Rotten Tomatoes"].astype(int)
combined_df["Runtime"]= combined_df["Runtime"].astype(int)


combined_df.dtypes
```
