```python
#Imports
import pandas as pd
import numpy as np
```

```python
#Read the csv files
production_df = pd.read_csv("ratings_integers_only.csv")
```

```python
#Delete other string columns
del production_df['Unnamed: 0']

production_df
```

```python
production=[]

for i in np.array(production_df['Production'].apply(lambda x: x.split(','))):
    for j in i:
        production.append(j.strip())
```

```python
set(production)
```

```python
# define columns first
for i in set(production):
    production_df[i]= pd.Series()

# populating colmuns 
for x in production_df['Production']:
    print(x.split(','))
```

```python
production_df['Miramax Films'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(production_df['Production']):
    
    for words in x.split(','):
        if words.strip() in set(production):
            production_df[words.strip()][i] = 1
```

```python
production_df.fillna(0, inplace=True)

del production_df['Production']
```

```python
production_df = production_df.apply(lambda x: pd.to_numeric(x), axis=0)
production_df = production_df.astype(int)

production_df.dtypes
```

```python
production_df.to_csv('production_integers_only.csv')
```
