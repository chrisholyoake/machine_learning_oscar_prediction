```python
from sklearn.datasets import make_blobs
# read file
Language_df = pd.read_csv("combined_with_genre_int.csv")

# how to convert pandas dataframe to a numpy array

# X, y = make_blobs(centers=2, random_state=42)

# print(f"Labels: {y[:10]}")
# print(f"Data: {X[:10]}")
```

```python
del Language_df['Unnamed: 0']
```

```python
Language_df.columns
```

```python
Language_df['Language']=Language_df['Language'].fillna("")
```

```python
language=[]

for i in np.array(Language_df['Language'].apply(lambda x: x.split(','))):
    for j in i:
        language.append(j.strip())
```

```python
set(language)
```

```python
# define columns first
for i in set(language):
    Language_df[i]= pd.Series()

# populating colmuns 
for x in Language_df['Language']:
    print(x.split(','))
```

```python
Language_df['English'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(Language_df['Language']):
    for words in x.split(','):
        if words.strip() in set(language):
            Language_df[words.strip()][i] = 1
```

```python
Language_df.fillna(0, inplace=True)

del Language_df['Language']

Language_df.columns
```

```python
Language_df = Language_df.apply(lambda x: pd.to_numeric(x), axis=0)
Language_df = Language_df.astype(int)

Language_df.dtypes
```
