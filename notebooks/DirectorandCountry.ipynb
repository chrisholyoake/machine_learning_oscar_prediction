```python
#Imports
import pandas as pd
import numpy as np
```

```python
#Read the csv files
director_df = pd.read_csv("olivia_csv_integers.csv")
director_df
```

```python
#Delete other string columns
del director_df['Unnamed: 0']
del director_df['Title']
del director_df['Country']
```

```python
director_df
```

```python
director=[]

for i in np.array(director_df['Director'].apply(lambda x: x.split(','))):
    for j in i:
        director.append(j.strip())
```

```python
set(director)
```

```python
# define columns first
for i in set(director):
    director_df[i]= pd.Series()

# populating colmuns 
for x in director_df['Director']:
    print(x.split(','))
```

```python
director_df['Quentin Tarantino'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(director_df['Director']):
    
    for words in x.split(','):
        if words.strip() in set(director):
            director_df[words.strip()][i] = 1
```

```python
director_df.fillna(0, inplace=True)

del director_df['Director']
```

```python
director_df = director_df.apply(lambda x: pd.to_numeric(x), axis=0)
director_df = director_df.astype(int)

director_df.dtypes
```

```python
director_df.to_csv('director_integers_only.csv')
```
